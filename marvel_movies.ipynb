# Marvel Movies

```python
# Importing necessary libraries
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

## Tasks
1. Scrape Iron Man movie info
2. Srape all marvel movies and save it in a file
3. Find the movie that has highest budget and box office 

### Task 1

```python
# Getting the movie wikipedia page
movie_url = 'https://en.wikipedia.org/wiki/Iron_Man_(2008_film)'
movie = requests.get(movie_url)
# Converting it into beautiulsoup object
soup = bs(movie.content)
```

```python
# Getting the infobox from wikipedia to get important information
def get_infovalue(data):
    if data.find('li') :
        return [li.get_text().replace('\xa0',' ') for li in data.find_all('li')]
    else:
        return data.get_text().replace('\xa0',' ')
    
    
movie_infobox = soup.select('.infobox.vevent tbody')
info_rows = movie_infobox[0].find_all("tr")
movie_info = {}
for index, movie in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = movie.get_text()
    elif index == 1:
        continue
    else:
        info_key = movie.find('th').get_text(' ', strip=True)
        info_value = get_infovalue(movie.find('td'))
        movie_info[info_key] = info_value
        
movie_info
```

### Task 2

```python
# URL to locate all marvel movies
movie_list_url = 'https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films'

movie_list = requests.get(movie_list_url)

bs_movie = bs(movie_list.content)

all_tables = bs_movie.select('.wikitable.plainrowheaders')
for tables in all_tables:
    if tables.find("div", {"class": "sortable"}):
        tables.decompose()
        
```

```python
final_table = all_tables[0].find_all('i') + all_tables[1].find_all('i') + all_tables[2].find_all('i')
len(final_table)
```

```python
# Function to get information from all the movies
movie_info_list = []
base_url = 'https://en.wikipedia.org'
def get_all_movies(url):
    movie = requests.get(url)
    soup = bs(movie.content)
    movie_infobox = soup.select('.infobox.vevent tbody')
    info_rows = movie_infobox[0].find_all("tr")
    movie_info = {}
    for index, movie in enumerate(info_rows):
        if index == 0:
            movie_info['title'] = movie.get_text()
        elif index == 1:
            continue
        else:
            info_key = movie.find('th').get_text(' ', strip=True)
            info_value = get_infovalue(movie.find('td'))
            movie_info[info_key] = info_value
    movie_info_list.append(movie_info)
    
for tab in final_table:
    get_all_movies(base_url + tab.a['href'])
    
movie_info_list
```

```python
# Saving the file for pandas analysis
import json
with open('marvel_movies.json', 'w', encoding='utf8') as output_file:
    json.dump(movie_info_list, output_file, ensure_ascii=False,indent=2)
```

### Task 3

```python
# Loading data for analysis
raw_data = pd.read_json('marvel_movies.json')
raw_data
```

```python
# Removing unwanted columns
raw_data = raw_data.drop(['Directed by','Screenplay by','Based on', 'Produced by','Starring','Cinematography','Edited by','Music by','Production company','Distributed by','Release date','Country','Language','Written by','Production companies','Story by'],axis=1)
```

```python
# Removing movies data that are yet to be released
raw_data = raw_data.drop(raw_data.index[24:34],0)
```

```python
raw_data.rename(columns={'Running time': 'Running time (mins)'},inplace=True)
raw_data['Running time (mins)'] = raw_data['Running time (mins)'].apply(lambda x: int(x.split('[')[0].replace('minutes','')))
```

```python
raw_data['Box office'] = raw_data['Box office'].apply(lambda x: x.split('[')[0])
```

```python
# Removing faulty data from row 9 and 10 to make it consistent dats
raw_data.at[9,'Budget'] = raw_data.iloc[9]['Budget'][1].split('(')[0]
raw_data.at[10,'Budget'] = raw_data.iloc[10]['Budget'][1].split('(')[0]
```

```python
import re
# regex expression to convert string into integer values for budget and box office
expression = r"\d+\.*\d*"
# Function to convert million and billion to numeric value
def word_to_digit_fig(num):
    value = num.split('[')[0]
    if 'million' in value:
        value = float(re.search(expression,value).group()) * 1000000.0
    elif 'billion' in value:
        value = float(re.search(expression,value).group()) * 1000000000.0
    else: 
        value = num
    return value

raw_data['Budget'] = raw_data['Budget'].apply(lambda x: word_to_digit_fig(x))
raw_data['Box office'] = raw_data['Box office'].apply(lambda x: word_to_digit_fig(x))
```

```python
raw_data
```

```python
# movie with longest running time
import matplotlib.pyplot as plt

def plotChart(factor,title,xlabel,ylabel):
    keys = [title for title, df in raw_data.groupby(['title'])]
    plt.figure(figsize=(15,5))
    plt.bar(raw_data['title'], raw_data[factor])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.xticks(keys, rotation='vertical', size=10)
    plt.show()
    
plotChart('Running time (mins)', 'Movie with longest running time', 'Movies', 'Running time (mins)')
```

```python
# movie with highest budget
plotChart('Budget', 'Movie with longest running time', 'Movies', 'Budget')
```

```python
# movie with highest box office
plotChart('Box office', 'Movie with longest running time', 'Movies', 'Box office')
```

```python
# budgest to box office comparison of marvel movies
ax = raw_data.plot(x="title", y="Box office", kind="bar")
raw_data.plot(x="title", y="Budget", kind="bar", ax=ax, color="C2")
plt.show()
```
